**La serie de Padovan es una sucesión de números naturales
definida por los valores iniciales P(0)=P(1)=P(2)=1, y la
relación de recurrencia para n ≥ 3:**  
**P n =P n−2 +P n−3)**  
**Implementar en pseudocódigo o en un lenguaje como
Python una función que calcule el término n-ésimo de la
serie de Padovan.**  

Los tres primeros términos de la serie son fijos:  
P(0)=1  
P(1)=1  
P(2)=1  

La serie se calcula utilizando una relación de recurrencia, que define cómo obtener cada término posterior en función de los términos anteriores:
P(n)=P(n−2)+P(n−3)

Cada término a partir del cuarto es la suma del término que está dos lugares antes y del que está tres lugares antes.

P(3)=P(1)+P(0)=1+1=2  
P(4)=P(2)+P(1)=1+1=2  
P(5)=P(3)+P(2)=2+1=3  
P(6)=P(4)+P(3)=2+2=4  
P(7)=P(5)+P(4)=3+2=5  
P(8)=P(6)+P(5)=4+3=7  
P(9)=P(7)+P(6)=5+4=9  
P(10)=P(8)+P(7)=7+5=12  
P(11)=P(9)+P(8)=9+7=16


**Seudocódigo:**  
Inicializar P(0) = 1  
Inicializar P(1) = 1  
Inicializar P(2) = 1  

Para n desde 3 hasta N hacer:
    P(n) = P(n-2) + P(n-3)

```python
#Solucion utilizando array
def padovan(N):
    # Los tres primeros términos de la serie de Padovan son fijos: 1, 1, y 1
    P = [1, 1, 1]

    # Calculamos iterativamente desde el índice 3 hasta N (incluido)
    for i in range(3, N):
        P.append(P[i-2] + P[i-3])  # Cada nuevo término es la suma de los términos en i-2 y i-3
    return P

while True:
    try:
        # Solicitamos al usuario que ingrese un número
        N = int(input("Número de términos? "))

        # Verificamos que el número sea positivo
        if N <= 0:
            print("Por favor, introduce un número entero positivo.")
        elif N == 1:
          print("Serie de Padovan:\n1")
          break
        elif N == 2:
          print("Serie de Padovan:\n1 1")
          break
        elif N == 3:
          print("Serie de Padovan:\n1 1 1")
          break
        else:
            print("Serie de Padovan:")
            P = padovan(N)
            print(P)
            break
    # Si la entrada no es valida
    except ValueError:
        print("Entrada no válida. Por favor, introduce un número entero.")
```
