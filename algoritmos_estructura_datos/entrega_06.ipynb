**• Considerar el siguiente algoritmo iterativo para
recorrer un árbol.**  
**1. Creamos una pila vacía S.**  
**2. Añadimos push el nodo actual a la pila. Y nos movemos al extremo izquierdo hasta llegar a extremo vacío.**  
**3. Si hemos llegado al extremo (y la pila no está vacía), mostramos el elemento superior de la pila. Y nos movemos a la derecha.**  
**4. Volvemos al paso 2.**  
**5. Finalizamos cuando el nodo está vacío y la pila está también vacía.**  

**• Describir para el árbol de nuestro ejemplo en
inorden: 4 2 5 1 3 , cada uno de los pasos del
algoritmo.**

```python
# Clase Nodo del árbol
class Node:
    # Constructor
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

# Recorrido del árbol
def inOrder(root):
    # Current empieza en la raíz
    current = root
    stack = []  # Creamos pila

    while True:
        # Nos movemos al nodo extremo izquierdo
        if current is not None:
            # Primero añadimos un puntero al nodo actual en la pila
            # Después nos movemos al hijo izquierdo
            stack.append(current)
            current = current.left
        # Backtrack desde el último nodo -vacío- y visitamos el nodo superior de la pila
        # (si está vacía, este proceso hijo izquierdo termina)
        elif stack:
            current = stack.pop()
            print(current.data, end=" ")
            # Hemos recorrido el nodo y su subárbol izquierdo
            # Recorremos ahora el derecho
            current = current.right
        else:
            break

    print()

# Ejemplo
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

inOrder(root)
```

**¿Qué es un árbol AVL?**

Es un árbol binario de búsqueda que se caracteriza por estar siempre balanceado.  
Fue desarrollado en 1962 por Adelson-Velsky y Landis (de ahí las siglas AVL).
Utiliza la altura de los nodos para monitorear su balance mediante el cálculo del factor de equilibrio.  

**Factor de Equilibrio:**
Es la diferencia entre la altura del subárbol derecho y la altura del subárbol izquierdo.  
Fórmula: Altura(Subárbol Derecho) - Altura(Subárbol Izquierdo).
Para que un árbol sea AVL, el factor de equilibrio de cualquier nodo debe ser **-1**, 0 o **1**.  

**Operaciones en un árbol AVL:**  
Las operaciones básicas como insertar, eliminar o buscar funcionan igual que en un árbol binario de búsqueda.
Después de insertar o eliminar un nodo, se verifica si el árbol sigue balanceado. Si no, se aplica una rotación para balancearlo.  

**Ejemplo de Árbol Degenerado vs. Árbol Balanceado:**  
Un árbol degenerado (similar a una lista enlazada) requiere más iteraciones para encontrar un nodo, lo que afecta el rendimiento.
Un árbol balanceado minimiza la profundidad, haciendo las búsquedas más eficientes.  

**Técnicas para balancear un árbol (Rotaciones):**  
Hay cuatro tipos de rotaciones:  
**Rotación Simple a la Derecha:**   
 Se aplica cuando un nodo está desbalanceado hacia la izquierda.  
**Rotación Simple a la Izquierda:**   
 Se aplica cuando un nodo está desbalanceado hacia la derecha.  
**Rotación Doble a la Derecha:**   
Primero, una rotación simple a la izquierda en el hijo izquierdo.
Luego, una rotación simple a la derecha en el nodo desbalanceado.  
**Rotación Doble a la Izquierda:**   

Primero, una rotación simple a la derecha en el hijo derecho.
Luego, una rotación simple a la izquierda en el nodo desbalanceado.  

**Ejemplo de Rotación Doble a la Derecha:**  
Paso 1: Aplicar una rotación simple a la izquierda en el subárbol izquierdo del nodo desbalanceado.  
Paso 2: Aplicar una rotación simple a la derecha en el nodo desbalanceado.
Resultado: El árbol queda balanceado, con factores de equilibrio de 0, 1 o -1 para todos los nodos.   

**Condiciones para aplicar cada rotación:**  
**Rotación Simple a la Derecha:**   
Nodo desbalanceado hacia la izquierda y su hijo izquierdo no está desbalanceado hacia la derecha.  
**Rotación Simple a la Izquierda:**   
Nodo desbalanceado hacia la derecha y su hijo derecho no está desbalanceado hacia la izquierda.  
**Rotación Doble a la Derecha:**   
Nodo desbalanceado hacia la izquierda y su hijo izquierdo está desbalanceado hacia la derecha.  
**Rotación Doble a la Izquierda:**   
Nodo desbalanceado hacia la derecha y su hijo derecho está desbalanceado hacia la izquierda.

Definición de Árbol Heap.  
Un árbol heap es un tipo de árbol binario completo que cumple con las siguientes propiedades:

Propiedad de Heap:  

**Max-Heap:**  
Cada nodo tiene un valor mayor o igual al de sus hijos. Es decir, el valor más grande está en la raíz.

**Min-Heap:**  
Cada nodo tiene un valor menor o igual al de sus hijos. Es decir, el valor más pequeño está en la raíz.

**Propiedad de Árbol Completo:**  

Es un árbol binario completo, lo que significa que todos los niveles del árbol están completamente llenos, excepto posiblemente el último nivel, que se llena de izquierda a derecha.
